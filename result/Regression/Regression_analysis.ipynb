```python
import json
import glob

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
files = ["log_max_max.dat", "log_max_average.dat", "log_max_sum.dat", "log_average_max.dat", "log_average_average.dat", "log_average_sum.dat", "log_sum_max.dat", "log_sum_average.dat", "log_sum_sum.dat"] 

neighbor_pooling = ["max", "average", "sum"]
graph_pooling = ["max", "average", "sum"]
color_list = ["#ff0000", "#aa0000", "#550000", "#00ff00", "#00aa00", "#005500", "#0000ff", "#0000aa", "#000055"]

plt.figure(figsize=(12, 8))
plt.rcParams["font.size"] = 24

for n_pool in neighbor_pooling:
    for g_pool in graph_pooling:
        df_log = pd.read_json("log_{0}_{1}.dat".format(n_pool, g_pool))
        print(df_log)
        plt.plot(range(1, 351), df_log["main/abs_max_loss"], label=("N:{0}\tG:{1}".format(n_pool[:3], g_pool[:3])).expandtabs())

plt.legend(title="Pooling Type".expandtabs(), bbox_to_anchor=(0.95, 0.5, 0.5, 0.4), borderaxespad=0.)
plt.show()

```

```python
plt.figure(figsize=(12, 8))
plt.rcParams["font.size"] = 24

for file in files:
    df_log = pd.read_json(file)
    n_pool = file[4:-4].split("_")[0][:3]
    g_pool = file[4:-4].split("_")[1][:3]
    
    plt.plot(range(1, 351), df_log["main/abs_loss"], label=("{0}\t{1}".format(n_pool, g_pool)).expandtabs())

plt.legend(title="PoolingType\tNeighbor\tGraph".expandtabs())
plt.show()
```

```python
plt.figure(figsize=(12, 8))

for file in glob.glob("*.dat"):
    df_log = pd.read_json(file)
    plt.plot(range(1, 351), df_log["validation/main/loss"], label=file[4:-4])
    print(file, len(df_log))

plt.ylim([0, 10])
plt.legend(title="Pooling Type".expandtabs(), bbox_to_anchor=(0.95, 0.5, 0.5, 0.4), borderaxespad=0.)
plt.show()
```
